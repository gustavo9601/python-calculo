## Función escalón o de Heaviside
$$H(x) =
     \begin{cases}
        0, &\quad \text{para, } x < 0 \\
        1,  &\quad\text{para. } x \ge 0 \\
     \end{cases}
$$

```python
def H(x):
  Y = np.zeros(len(x))
  for idx,x in enumerate(x):
    if x>=0:
      Y[idx]=1
  return Y


N=1000


y = H(x)

plt.plot(x,y)
plt.grid()
```

## Función sigmoide

$$f(x)=\frac{1}{1-e^{-x}}$$

Va del rango entre 0 y 1

```python
def f_sigmo(x):
  return 1/(1 + np.exp(-x))


N=1000

y = f_sigmo(x)

plt.plot(x,y)
plt.grid()
```

## Función tangente hiperbólica

- Funcion de escalamiento
- Va del rango entre 0 y 1

$$f(x)=\frac{2}{1+e^{-2x}}-1$$


```python
def f(x):
  return np.tanh(x)


N=1000

y = f(x)

plt.plot(x,y)
plt.grid()
```

## Función ReLU

- Si x <= 0, entonces 0 en caso contrario el valor de x
-
$$R(x)=max(0,x)$$

```python
def f(x):
  return np.maximum(x,0)


N=1000

y = f(x)

plt.plot(x,y)
plt.grid()
```
